```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scenarios import RockCreator as RC
```

```python
data = "Force2020_Well_159-14.csv"
```

```python
df_force2020 = pd.read_csv(data, sep=";", header=0, na_values='.')
```

```python
df_force2020.describe()
```

```python
fig, axs = plt.subplots(4, 2, figsize=(15,20))
inputs = ["RHOB", "GR", "NPHI", "PEF", "DTC", "SP", "RMED"]
color = "#1f77b4"
edgecolor = "black"
for i in range(0, 4):
    for j in range(0, 2):
        axs[i,j].set_facecolor("whitesmoke")
        axs[i,j].grid(color="black", linestyle='dashed')
        axs[i,j].set_ylabel("Frequency")
        axs[i,j].set_axisbelow(True)
axs[0,0].hist(df_force2020[inputs[0]], bins=25, color=color, edgecolor=edgecolor)
axs[0,0].set_xlabel("RHOB (g/cm$^3$)")
axs[0,1].hist(df_force2020[inputs[1]], bins=25, color=color, edgecolor=edgecolor)
axs[0,1].set_xlabel("GR (API)")
axs[1,0].hist(df_force2020[inputs[2]], bins=25, color=color, edgecolor=edgecolor)
axs[1,0].set_xlabel("NPHI")
axs[1,1].hist(df_force2020[inputs[3]], bins=25, color=color, edgecolor=edgecolor)
axs[1,1].set_xlabel("PEF")
axs[2,0].hist(df_force2020[inputs[4]], bins=25, color=color, edgecolor=edgecolor)
axs[2,0].set_xlabel("DTC")
axs[2,1].hist(df_force2020[inputs[5]], bins=25, color=color, edgecolor=edgecolor)
axs[2,1].set_xlabel("SP")
axs[3,0].hist(df_force2020[inputs[6]], bins=25, color=color, edgecolor=edgecolor)
axs[3,0].set_xlabel("RMED")
#
```

```python
fig, ax = plt.subplots(2, 2, dpi=100, figsize=(8,16))
ax[0,0].plot(df_force2020["RHOB"], df_force2020["DEPTH_MD"])
ax[0,0].invert_yaxis()
ax[0,1].plot(df_force2020["GR"], df_force2020["DEPTH_MD"])
ax[0,1].invert_yaxis()
ax[1,0].plot(df_force2020["PEF"], df_force2020["DEPTH_MD"])
ax[1,0].invert_yaxis()
ax[1,1].plot(df_force2020["DTC"], df_force2020["DEPTH_MD"])
ax[1,1].invert_yaxis()
```
