```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from modules.minerals import feldspars
```

```python
data = feldspars()
```

```python
data_kfs = data.alkalifeldspar(keyword="None")
```

```python
print(data_kfs)
```

```python
kfs_data = []
for i in range(1000):
    data_kfs = data.alkalifeldspar(keyword="None")
    kfs_data.append([data_kfs[0], data_kfs[1][0], data_kfs[2], data_kfs[4][0], data_kfs[4][1], data_kfs[5][0], data_kfs[5][2], data_kfs[6][1], data_kfs[6][4]])
```

```python
#print(kfs_data)
your_list = kfs_data
df = pd.DataFrame (your_list, columns=["NAME", "M", "RHOB", "VP", "VS", "GR", "U", "Na", "K"])
```

```python
df.round(decimals=2)
```
