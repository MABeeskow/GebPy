```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = "../outputs/RockProperties_Granite.csv"
```

```python
df = pd.read_csv(data, sep=',', header=0, na_values='.')
```

```python
df
```

```python
df.describe()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15,20))
inputs = ["RHOB", "VP", "VS", "GR", "PHIN", "POISSON"]
for i in range(0, 3):
    for j in range(0, 2):
        for k in range(0, len(inputs)):
            axs[i,j].grid(color="#000000", linestyle='dashed')
            axs[i,j].set_ylabel("Frequency")
            axs[i,j].set_axisbelow(True)
axs[0,0].hist(df[inputs[0]], bins=25, color="#00549F", edgecolor="#000000")
axs[0,0].set_xlabel("RHOB (g/cm$^3$)")
axs[0,1].hist(df[inputs[3]], bins=25, color="#00549F", edgecolor="#000000")
axs[0,1].set_xlabel("GR (API)")
axs[1,0].hist(df[inputs[1]], bins=25, color="#00549F", edgecolor="#000000")
axs[1,0].set_xlabel("VP (m/s)")
axs[1,1].hist(df[inputs[2]], bins=25, color="#00549F", edgecolor="#000000")
axs[1,1].set_xlabel("VS (m/s)")
axs[2,0].hist(df[inputs[4]], bins=25, color="#00549F", edgecolor="#000000")
axs[2,0].set_xlabel("PHIN (%)")
axs[2,1].hist(df[inputs[5]], bins=25, color="#00549F", edgecolor="#000000")
axs[2,1].set_xlabel("POISSON (1)")
plt.grid(color="#000000", linestyle='dashed')
#
```
