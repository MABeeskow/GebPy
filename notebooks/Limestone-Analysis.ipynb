```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataRP = "../outputs/RockProperties_Limestone.csv"
dataGC = "../outputs/Geochemistry_Limestone.csv"
```

```python
dfRP = pd.read_csv(dataRP, sep=',', header=0, na_values='.')
dfGC = pd.read_csv(dataGC, sep=',', header=0, na_values='.')
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15,20))
inputs = ["RHOB", "VP", "VS", "GR", "PHIN", "POISSON"]
for i in range(0, 3):
    for j in range(0, 2):
        for k in range(0, len(inputs)):
            axs[i,j].grid(color="#000000", linestyle='dashed')
            axs[i,j].set_ylabel("Frequency")
            axs[i,j].set_axisbelow(True)
axs[0,0].hist(dfRP[inputs[0]], bins=25, color="#00549F", edgecolor="#000000")
axs[0,0].set_xlabel("RHOB (g/cm$^3$)")
axs[0,1].hist(dfRP[inputs[3]], bins=25, color="#00549F", edgecolor="#000000")
axs[0,1].set_xlabel("GR (API)")
axs[1,0].hist(dfRP[inputs[1]], bins=25, color="#00549F", edgecolor="#000000")
axs[1,0].set_xlabel("VP (m/s)")
axs[1,1].hist(dfRP[inputs[2]], bins=25, color="#00549F", edgecolor="#000000")
axs[1,1].set_xlabel("VS (m/s)")
axs[2,0].hist(dfRP[inputs[4]], bins=25, color="#00549F", edgecolor="#000000")
axs[2,0].set_xlabel("PHIN (%)")
axs[2,1].hist(dfRP[inputs[5]], bins=25, color="#00549F", edgecolor="#000000")
axs[2,1].set_xlabel("POISSON (1)")
plt.grid(color="#000000", linestyle='dashed')
#
```

```python
from pandas.plotting import scatter_matrix
import seaborn as sn
```
