```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataRP = "../outputs/RockProperties_Sandstone.csv"
```

```python
dfRP = pd.read_csv(dataRP, sep=',', header=0, na_values='.')
```

```python
dfRP
```

```python
dfRP.describe()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15,20))
inputs = ["RHOB", "VP", "VS", "GR", "PHIN", "POISSON"]
color = "#1f77b4"
edgecolor = "black"
for i in range(0, 3):
    for j in range(0, 2):
        for k in range(0, len(inputs)):
            axs[i,j].set_facecolor("whitesmoke")
            axs[i,j].grid(color="black", linestyle='dashed')
            axs[i,j].set_ylabel("Frequency")
            axs[i,j].set_axisbelow(True)
axs[0,0].hist(dfRP[inputs[0]], bins=25, color=color, edgecolor=edgecolor)
axs[0,0].set_xlabel("RHOB (g/cm$^3$)")
axs[0,1].hist(dfRP[inputs[3]], bins=25, color=color, edgecolor=edgecolor)
axs[0,1].set_xlabel("GR (API)")
axs[1,0].hist(dfRP[inputs[1]], bins=25, color=color, edgecolor=edgecolor)
axs[1,0].set_xlabel("VP (m/s)")
axs[1,1].hist(dfRP[inputs[2]], bins=25, color=color, edgecolor=edgecolor)
axs[1,1].set_xlabel("VS (m/s)")
axs[2,0].hist(dfRP[inputs[4]], bins=25, color=color, edgecolor=edgecolor)
axs[2,0].set_xlabel("PHIN (%)")
axs[2,1].hist(dfRP[inputs[5]], bins=25, color=color, edgecolor=edgecolor)
axs[2,1].set_xlabel("POISSON (1)")
#
```
