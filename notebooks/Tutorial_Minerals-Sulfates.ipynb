```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from matplotlib.collections import PathCollection
import seaborn as sns
sns.set()
from modules.sulfates import Sulfates
```

```python
print("TEST - SULFATES (from sulfates.py)")

data_anh = Sulfates(impurity="pure").create_anhydrite()
data_gp = Sulfates(impurity="pure").create_gypsum()
data_sch = Sulfates(impurity="pure").create_scheelite()
data_bar = Sulfates(impurity="pure").create_barite()
data_cls = Sulfates(impurity="pure").create_celestine()
data_ang = Sulfates(impurity="pure").create_anglesite()
data_han = Sulfates(impurity="pure").create_hanksite()
data_clc = Sulfates(impurity="pure").create_chalcanthite()
data_kie = Sulfates(impurity="pure").create_kieserite()
data_jar = Sulfates(impurity="pure").create_jarosite()
data_al = Sulfates(impurity="pure").create_alunite()

data_sulfates = [data_anh, data_gp, data_sch, data_bar, data_cls, data_ang, data_han, data_clc, data_kie, data_jar, data_al]

data_reduced = []
for item in data_sulfates:
    data_reduced.append([item[0], item[1], item[2], item[3][0], item[3][1], item[3][3], item[4][0], item[4][1], item[4][2], item[5][0], item[5][1], item[5][2], item[-1]])
```

```python
df_sulfates = pd.DataFrame(data_reduced, columns=["NAME", "M", "RHOB", "K", "G", "POISSON", "VP", "VS", "VP/VS", "GR", "PE", "U", "SYSTEM"])
```
